```python
plt.figure(figsize=(8, 6))
plt.scatter(data=df,x ='TrafficIndexLive', y= 'Weather_Condition',s=30, c='green')
plt.title(' the relationship between TrafficIndexLive and Date')
plt.xlabel('TrafficIndexLive')
plt.ylabel('Weather_Condition')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(data=df,x ='JamsCount', y= 'TrafficIndexLive',s=30, c='green')
plt.title(' the relationship between Jams Count and Traffic Index Live')
plt.xlabel('Jams Count')
plt.ylabel('Traffic Index Live')
plt.show()
```

```python
#I use bar charts
# Jams Count:Number of traffic jams. city
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='JamsCount', y='City', palette='muted')
plt.xlabel('JamsCount')
plt.ylabel('City')
plt.xticks(rotation=90)
plt.show()
```
